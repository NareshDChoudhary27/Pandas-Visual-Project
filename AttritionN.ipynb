```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming the DataFrame is already read and stored in a variable `df`
df = pd.read_csv('/content/HR_Analytics (1)Recent.csv')

# 1. Bar Plot for Attrition by  SalarySlab
plt.figure(figsize=(12, 6))
sns.barplot(x='AttritionCount', y='SalarySlab', data=df, ci=None)
plt.xticks(rotation=90)
plt.title('Attrition by  SalarySlab')
plt.xlabel('AttritionCount')
plt.ylabel('SalarySlab')
plt.show()
```

```python
# 2. Bar Plot for Attrition by Age
plt.figure(figsize=(12, 6))
sns.barplot(x='AgeGroup', y='AttritionCount', data=df, ci=None)
plt.xticks(rotation=90)
plt.title('Attrition by Age')
plt.xlabel('AgeGroup')
plt.ylabel('AttritionCount')
plt.show()
```

```python
# 3. Line Plot for Attrition by job role
plt.figure(figsize=(12, 6))
sns.lineplot(x='JobRole', y='AttritionCount', data=df, marker='o')
plt.title(' Attrition by job role')
plt.xlabel('JobRole')
plt.ylabel('AttritionCount')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# 4. Pie Chart for Attrition by Education
attrition_by_education = df.groupby('EducationField')['AttritionCount'].sum() # Count Attrition by EducationField
attrition_by_education.plot.pie(autopct='%1.1f%%', figsize=(8, 8), title='Attrition by Education')
plt.ylabel('AttritionCount')
plt.show()
```

```python
#5.Area plot for  Attrition by Years at Company
attrition_by_years = df.groupby('YearsAtCompany')['AttritionCount'].mean()

plt.figure(figsize=(10, 6))
attrition_by_years.plot(kind='area', color='Red') # Removed extra .plot
plt.title('Attrition Rate Based on Years at Company')
plt.xlabel('YearsAtCompany')
plt.ylabel('AttritionCount')
plt.ylim(0, 1)
plt.grid(True)
plt.show()
```

```python
# 6. Bar Plot for Attrition by Gender
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='AttritionCount', data=df, ci=None)
plt.xticks(rotation=90)
plt.title('Attrition by Gender')
plt.xlabel('Gender')
plt.ylabel('AttritionCount')
plt.show()
```
